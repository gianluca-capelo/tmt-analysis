```python
from src import config
import pandas as pd

df = pd.read_csv(config.METADATA_CSV)
df = df[df['Nro de participante'] != 'Control = ']
df = df[df['Nro de participante'] != 'Caso = ']
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume your DataFrame is named 'df'
# Compute descriptive statistics for numerical columns
desc_stats = df.describe()
print("Descriptive Statistics:")
print(desc_stats)

# Define key variables you want to visualize
key_vars = ['Edad', 'MMSE / 30', 'Reloj / 3']

# Plot histograms for each key variable to view their distribution
for var in key_vars:
    plt.figure()  # Create a new figure for each plot
    df[var].hist(bins=15)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()

# Create boxplots for the key variables to check for outliers
plt.figure()
df[key_vars].boxplot()
plt.title('Boxplots for key variables')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x='Género', data=df)
plt.title('Cantidad de Participantes por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Género', y='Edad', data=df, inner='quartile')
plt.title('Violin Plot of Age by Gender')
plt.xlabel('Gender (F/M)')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Boxplot for Age by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Género', y='Edad', data=df)
plt.title('Distribution of Age by Gender')
plt.xlabel('Gender (F/M)')
plt.ylabel('Age')
plt.show()

# Example: Boxplot for MMSE score by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Género', y='MMSE / 30', data=df)
plt.title('Distribution of MMSE / 30 by Gender')
plt.xlabel('Gender (F/M)')
plt.ylabel('MMSE / 30')
plt.show()
```
